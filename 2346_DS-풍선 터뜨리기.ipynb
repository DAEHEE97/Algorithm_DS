## memo

- 풍선 고유 번호 저장 구조

- 현재 문제는 원형으로 놓여 있는 풍선을 대상으로 하고, 왼쪽, 오른쪽 양방향으로 이동이 가능해야 하므로 deque.rotate() 처리
    - 양수 음수에 따라, 회전 방향과, 칸수 처리


---

## deque rotate

- deuqe 시계방향 원 데이터

- rotate 양수, 원형 시계방향 회전, > 화살표 왼쪽으로 이동
- rotate 음수, 원형 반시계 방향 회전, > 화살표 오른쪽으로 이동

- pop, rotate 양수, pop 처리 후, 화살표 오른쪽으로 한칸 이동 > 전 위치에서 이동하는 것을 고려 > v-1 칸 이동
- pop, rotate 음수, pop 처리 후, 화살표 오른쪽으로 한칸 이동 > 어차피 왼쪽으로 이동하는 칸수 같음 v 칸 이동


---

```python
from collections import deque

N = int(input())

# 고유 번호 저장 > enumerate를 이용한, (i,v) tuple 요소 데이터 형
q = deque([(i+1, v) for i, v in enumerate(map(int, input().split()))])

ans = []

while q:
    
    n, v = q.popleft()
    ans.append(n)
    
    # pop 처리시 양수, 오른쪽으로 이동, pop 이후, 화살표 이동 > v-1 이동
    if v > 0:
        
        q.rotate(-(v-1))
        
    
    # pop 처리시 음수, 왼쪽 이동, pop 이후, 상괎없이 v 이동
    elif v < 0:
        
        q.rotate(-v)
        
print(*ans)
```
